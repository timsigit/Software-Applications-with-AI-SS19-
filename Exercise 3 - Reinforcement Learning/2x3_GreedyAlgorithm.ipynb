```python
import numpy as np


with open('SAKI Exercise 3 warehouseordernew.txt') as f:
    lineList = f.readlines()

testdata = np.empty([len(lineList), 2], dtype=str)
for idx, line in enumerate(lineList):
    words = line.split()
    testdata[idx, 0] = words[0]
    testdata[idx, 1] = words[1]

warehouse = ['0', '0', '0', '0', '0', '0']
```

```python
distance = 0  # Calculate distance for later evaluation
for transIdx, transition in enumerate(testdata):
    if transIdx < len(testdata) - 1:
        trans = transition[0]
        col = transition[1]

        if col == 'w':
            col = '1'
        elif col == 'b':
            col = '2'
        else:
            col = '3'

        for idx, pos in enumerate(warehouse):
            if trans == 's':
                if idx == 0:
                    if warehouse[idx] == '0':
                        warehouse[idx] = col
                        distance += 2
                        break
                elif idx == 1 or idx + 2 == 3:
                    if warehouse[idx] == '0':
                        warehouse[idx] = col
                        distance += 4
                        break
                    if warehouse[idx + 2] == '0':
                        warehouse[idx + 2] = col
                        distance += 4
                        break
                elif idx == 2 or idx + 2 == 4:
                    if warehouse[idx] == '0':
                        warehouse[idx] = col
                        distance += 6
                        break
                    if warehouse[idx + 2] == '0':
                        warehouse[idx + 2] = col
                        distance += 6
                        break
                else:
                    if warehouse[idx] == '0':
                        warehouse[idx] = col
                        distance += 8
                        break
            else:
                if idx == 0:
                    if warehouse[idx] == col:
                        warehouse[idx] = '0'
                        distance += 2
                        break
                elif idx == 1 or idx + 2 == 3:
                    if warehouse[idx] == col:
                        warehouse[idx] = '0'
                        distance += 4
                        break
                    if warehouse[idx + 2] == col:
                        warehouse[idx + 2] = '0'
                        distance += 4
                        break
                elif idx == 2 or idx + 2== 4:
                    if warehouse[idx] == col:
                        warehouse[idx] = '0'
                        distance += 6
                        break
                    if warehouse[idx + 2] == col:
                        warehouse[idx + 2] = '0'
                        distance += 6
                        break
                else:
                    if warehouse[idx] == col:
                        warehouse[idx] = '0'
                        distance += 8
                        break


        print('------')


        mapping_matrix = {
            "0": '-',
            "1": 'W',
            "2": 'B',
            "3": 'R'
        }

        print(' _____ _____ _____')
        print('|     |     |     |')
        print('| ', mapping_matrix[warehouse[5]], ' | ', mapping_matrix[warehouse[4]], ' | ', mapping_matrix[warehouse[3]],
              ' |')
        print('|_____|_____|_____|')
        print('|     |     |     |')
        print('| ', mapping_matrix[warehouse[2]], ' | ', mapping_matrix[warehouse[1]], ' | ', mapping_matrix[warehouse[0]],
              ' |')
        print('|_____|_____|_____|', ' (1,1)')

        print('')
        print(warehouse)
        print('')



print('The robot has covered the following distance: ', distance)
print('')
```
